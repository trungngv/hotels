## Room capacities

### Total room capacity

```python
hn = df_hn.roomCapacity.sum()
nt = df_nt.roomCapacity.sum()
pq = df_pq.roomCapacity.sum()
print 'Ha Noi: ', hn
print 'Nha Trang: ', nt
print 'Phu Quoc: ', pq

plt.bar((1, 2, 3), (hn, nt, pq), align='center')
plt.xticks((1, 2, 3), ('Hanoi', 'Nha Trang', 'Phu Quoc'))
plt.ylabel('Total number of rooms')
plt.title('Comparison of room capacities')
```

### Distribution
We can divide hotels into 3 categories based on their capacity: small (0 - 50 rooms), medium (50 - 100 rooms), large (over 100 rooms). There are a few hotels in all regions with very large capacity (over 300 rooms), these are exceptions and we ignore in this analysis. 

It's clear from the distribution that most hotels in Hanoi and Nha Trang are small. Although in Nha Trang the market for medium to large hotels is bigger.  

Hypothesis: land is less expensive in Nha Trang than HN; hotels in Nha Trang more spread out.

```python
from scipy import stats

plt.figure(figsize = (15, 5))

plt.subplot(131)
out = plt.hist(df_hn.roomCapacity[df_hn.roomCapacity < 300], bins = 30)
plt.xlabel('Hotel capacity')
plt.ylabel('Count')
plt.title('Hotel capacities in Hanoi')

plt.subplot(132)
plt.hist(df_nt.roomCapacity[df_nt.roomCapacity < 300], bins = 30)
plt.xlabel('Hotel capacity')
plt.ylabel('Count')
plt.title('Hotel capacities in Nha Trang')

plt.subplot(133)
plt.hist(df_pq.roomCapacity[df_pq.roomCapacity < 300], bins = 15)
plt.xlim([0, 300])
plt.xlabel('Hotel capacity')
plt.ylabel('Count')
plt.title('Hotel capacities in Phu Quoc')
```

## Hotel grade

```python
def convert_rating(x):
    if ' hotel' in x:
        return x.split()[0]
    elif 'stars' in x:
        return "%s-star" % x.split()[0]

def get_ratings(df):
    ratings = df.starRating.astype(np.string_)
    ratings = ratings[ratings != '']
    return ratings.map(convert_rating)
```

### Percentage of star-ratings

Consistent with our analysis on the hotel capacity, 2-star and 3-star hotels dominate all cities. There are still many 1-star hotels in both Nha Trang and Phu Quoc, so it'll be interesting to see how that develops into the future. My prediction is that the number of 1-star hotels in Nha Trang, driving the distribution to be more or less equivalent to Hanoi. 

Opportunity: find 1-star hotels with very low rating (in crowded area) and buy them, then upgrade


```python
plt.figure(figsize = (15, 5))

ratings = get_ratings(df_hn).value_counts().sort_index()
plt.subplot(131)
plt.pie(ratings.values, labels = ratings.index, autopct = '%.2f')
plt.title('Ha noi')

ratings = get_ratings(df_nt).value_counts().sort_index()
plt.subplot(132)
plt.pie(ratings.values, labels = ratings.index, autopct = '%.2f')
plt.title('Nha Trang')

ratings = get_ratings(df_pq).value_counts().sort_index()
plt.subplot(133)
plt.pie(ratings.values, labels = ratings.index, autopct = '%.2f')
plt.title('Phu Quoc')
```

```python
def plot_review_scores(df, sp):
    scores = df.ix[df.avgReviewScore != '', 'avgReviewScore'].astype(np.float32)
    plt.subplot(sp)
    plt.hist(scores)
    plt.xlabel('avg. review scores')
    return scores

plt.figure(figsize = (15, 5))
plot_review_scores(df_hn, 131)
plt.title('Hanoi')

plot_review_scores(df_nt, 132)
plt.title('Nha Trang')

plot_review_scores(df_pq, 133)
plt.title('Phu Quoc')

```

Hotels in Hanoi are rated higher on average, indicating that the market is more competitive there. There are still opportunities for Nha Trang and Phu Quoc.

## Guest nationalities

This can be biased toward English-speaking countries because some website only show reviews based on the viewer's location.



```python
from collections import Counter

def to_int(xs):
    return [int(x) for x in xs]

def get_nation_counts(df):
    c = Counter()
    for (i, names) in enumerate(df.topNationNames):
        counts = df.topNationCounts.values[i]
        counts = counts.replace('[', '').replace(']','')
        d = dict(zip(names.split(','), to_int(counts.split(','))))
        c.update(d)
    return c

def plot_nation_counts(df, top=20):
    counter = zip(*(get_nation_counts(df).most_common(top)))
    pos = np.arange(len(counter[0]), 0, -1)
    plt.barh(pos, counter[1], align='center')
    plt.yticks(pos, counter[0])

plt.figure()    
plot_nation_counts(df_hn, 20)
plt.figure()
plot_nation_counts(df_nt, 20)
plt.figure()
plot_nation_counts(df_pq, 20)
```

## Location

```python
#plt.figure(figsize = (15, 5))

def color_map(score):
    if score != '':
        return float(score) / 10
    else:
        return 0
plt.figure(figsize = (20, 16))    
plt.scatter(df_hn.latitude, df_hn.longitude, s=df_hn.roomCapacity,
            c=df_hn.avgReviewScore.map(color_map))
plt.title('Ha Noi')
plt.colorbar()
```

```python
plt.figure(figsize = (15, 16))
plt.scatter(df_nt.latitude, df_nt.longitude, s=df_nt.roomCapacity,
            c=df_nt.avgReviewScore.map(color_map))
plt.title('Nha Trang')
plt.colorbar()
```

```python
plt.figure(figsize = (15, 16))
plt.scatter(df_pq.latitude, df_pq.longitude, s=2*df_pq.roomCapacity,
            c=df_pq.avgReviewScore.map(color_map))
plt.title('Phu Quoc')
plt.colorbar()
```
